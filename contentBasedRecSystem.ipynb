```python
# function to create a sinle vector for all users seen movies
def user_vector(ratings,movie_vector):
    vector=[]
    for user in range(1,944): # Each user id to create vector
        u1=ratings[ratings['user_id']==user][['movie_id','rating']]  
        u1_list=[]
        for i in u1['movie_id']: # Each user watched movie genre vector
            f=movie_vector[i].values
            u1_list.append(f)
        table_u1=pd.DataFrame(u1_list,columns=range(1,19))
        vec=[]
        for j in range(1,19): # Sum up the each genre of the user
            g=sum(table_u1[j])
            vec.append(g/len(u1))
        vector.append(vec)
    return vector
```

```python
vector=user_vector(ratings,movie_vector)
```

```python
# converting into Dataset
users_vector=pd.DataFrame(vector)
```

```python
# Dot matrix for  user_vector and movie_vetcor
content_vector=np.dot(users_vector,movie_vector)
```
